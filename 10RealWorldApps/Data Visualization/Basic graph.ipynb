```python
#Making a basic Bokeh line graph

#importing Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data for line
x = [1,2,3,4,5]
y = [6,7,8,9,10]

#prepare the output file
output_file("line.html")

#create figure object
f = figure()

#create line plot
f.line(x,y)

#showing figure object
show(f)
```

```python
#Making a basic Bokeh line graph

#importing Bokeh and Pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data for line
df = pandas.read_csv("http://pythonhow.com/data/bachelors.csv")
x = df["Year"]
y = df["Engineering"]

#prepare the output file
output_file("line_from_csv_url.html")

#create figure object
f = figure(plot_width=1000, plot_height=900, tools='pan')
f.title.text = "Some Cool Data"
f.title.text_color = "Gray"
f.title.text_font_style = "bold"
f.xaxis.minor_tick_line_color = None
f.yaxis.minor_tick_line_color = None
f.xaxis.axis_label = "Date"
f.yaxis.axis_label = "Intensity"

#create line plot
f.line(x,y)

#showing figure object
show(f)
```

```python
#Making a basic Bokeh line graph

#importing Bokeh and Pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data for line
df = pandas.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx", sheet_name=0)
df["Temperature"] = df["Temperature"] / 10
df["Pressure"] = df["Pressure"] / 10

#prepare the output file
output_file("line_from_excel.html")

#create figure object
f = figure(plot_width=500, plot_height=400, tools='pan')
f.title.text = "Temperatore and Air Pressure"
f.title.text_color = "Gray"
f.title.text_font_style = "bold"
f.xaxis.minor_tick_line_color = None
f.yaxis.minor_tick_line_color = None
f.xaxis.axis_label = "Temperature (â„ƒ)"
f.yaxis.axis_label = "Pressure (hPa)"

#create line plot
f.circle(df["Temperature"],df["Pressure"], size=0.5)

#showing figure object
show(f)
```

```python
#Making a basic Bokeh line graph

#importing Bokeh and Pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.models import HoverTool
import pandas

#prepare some data for line
x = [1,2,3,4,5]
y = [2,6,10,5,3]

#prepare the output file
output_file("Scatter_plotting.html")

#create figure object
f = figure(plot_width=1000, plot_height=900, tools='pan,reset')
f.title.text = "Earthquake"
f.title.text_color = "Orange"
f.title.text_font_style = "italic"
f.yaxis.minor_tick_line_color = "Yellow"
f.xaxis.axis_label = "Times"
f.yaxis.axis_label = "Value"
hover = HoverTool(tooltips=[("Time", "@x"),("Value", "@y")])

f.add_tools(hover)

#create line plot
f.circle(x, y, size=[i*5 for i in y], color=["red" if i>4 else "green" for i in y], alpha=0.5)
for a in x:
    f.line(a,y[x.index(a)], color="red")

#showing figure object
show(f)
```
